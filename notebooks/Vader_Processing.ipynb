```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()
```

```python
import pandas as pd
# Load the preprocessed data
df = pd.read_csv("../data/processed_twitter_data.csv")
```

```python
# Get Scores
def analyze_sentiment(text):
    if pd.notna(text):
        return analyzer.polarity_scores(text)
    return None

df['sentiment_dict'] = df['processed_text'].apply(analyze_sentiment)
```

```python
# Save processed data
df.to_csv("../data/processed_tweets.csv", index=False)
print("✅ Processed data saved to: ../data/processed_tweets.csv")
```
